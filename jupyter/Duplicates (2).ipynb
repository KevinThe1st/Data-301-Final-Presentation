# Duplicate lines top 100

**Here we calculated the number of duplicate lines in the lyrics for each song. To do this we found the number of total lines and subtracted from it the number of unique lines.**

```python
import pickle
import pandas as pd
import numpy as np
import seaborn as sns
import random
%matplotlib inline
```

```python
top100 = pickle.load(open("top100_df.pkl", "rb"))
```

```python
lyrics = top100.lyrics
lyrics
```

**Cleaning and calculation.**

```python
duplicates = []
for i in range(0, len(lyrics)):
    try:
        l = [x for x in lyrics.loc[i].split("\n") if x]
        tot = len(l)
        nonRepeat = list(set(l))
        length_set = len(nonRepeat)
        dups = tot-length_set
        duplicates.append(dups)
    except:
        duplicates.append(0)
```

**Inserting the new variable into the data frame.**

```python
top100["num_dupes"] = duplicates
```
