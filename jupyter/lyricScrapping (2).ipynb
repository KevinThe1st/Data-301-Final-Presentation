**To do this we had to scrape from `wikia lyrics`, but had to obtain the lyrics in multiple ways. This reason for this being is that the songs are fairly new. We still were not able to get the lyrics for 22 songs.**

```python
from bs4 import BeautifulSoup
import requests
```

**From the `wiki lyrics` website we were able to obtain the links to the pages of the lyrics since each song had a link in the form of `http://lyrics.wikia.com/wiki/ARTIST:TITLE`.**

**Creating the links.**

```python
lyric_links = []
#http://lyrics.wikia.com/wiki/Jonas_Brothers:Sucker
for i in range(0, 100):
    artist = df.artist.loc[i].replace(" ", "_")
    title = df.title.loc[i].replace(" ", "_")
    link = "http://lyrics.wikia.com/wiki/" + artist + ":" + title    
    lyric_links.append(link)
```

**Extracting lyrics from each link.**

```python
lyrics = []
for i in range(0, 100):
    resp = requests.get(lyric_links[i])
    soup = BeautifulSoup(resp.content, "html.parser")
    
    try:
        lyric = (soup.find("div", class_="lyricbox").decode_contents()
              .replace('<div class="lyricsbreak"></div>', "")
              .replace("<br/>", "\n"))
        lyrics.append(lyric)
    except:
        lyrics.append(np.nan)
    
```

```python
df["lyrics"] = lyrics
```
